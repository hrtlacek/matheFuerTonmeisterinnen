```python
import sympy as sp
```

```python
π = pi
sr = 44100
T = 5
N = int(5*sr)
t = arange(N)/sr
```

```python
fs = array([[100,200,300,400]]).T
x = sin(2*fs*t*π)
```

```python
plot(t, x.T)
xlim([0, 1/100])
```

```python
x.shape
```

```python
A = zeros([6,4])
A[0,0] = 1
A[1,1] = 1
A[2,2] = 1
A[3,3] = 1
A[4,0] = 1
A[5,0] = 1
```

```python
sp.Matrix(A)
```

```python
y = A@x
```

```python
y.shape
```

```python
plot(t, y.T)
xlim([0, 1/100])
```
