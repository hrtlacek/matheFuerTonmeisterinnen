# Geometrische Interpretation eines Linearen Gleichungssystems

Gegeben:

$$ 2x +3y = 8$$
$$ 4x - y = 2$$

```python
%pylab inline
import sympy as sp
from sympy.abc import x,y
sx = 6 # plot groesse x.
sy = 3 # plot groesse y.
```

```python
eq1 = sp.Eq(2*x + 3*y,8)
eq1
```

```python
eq2 = sp.Eq(4*x - y,2)
eq2
```

```python
solution1 = sp.solve(eq1, y)
solution2 = sp.solve(eq2, y)
```

```python
solution1[0]
```

```python
solution2[0]
```

```python
y1 = sp.lambdify(x,solution1[0])
y2 = sp.lambdify(x,solution2[0])
```

```python
figure(figsize=[sx,sy])

x = linspace(-4,4, 200)
y = y1(x)
plot(x,y, label='$'+sp.latex(solution1[0])+'$')
y = y2(x)
plot(x,y, label='$'+sp.latex(solution2[0])+'$')


A = array([[2,3],[4,-1]])
b = array([8,2])
sol = np.linalg.solve(A,b)
scatter(*sol,c='r', label=f'LÃ¶sung: $x={sol[0]}$, $y={sol[1]}$')

xlabel('$x$')
ylabel('$y$')

ylim([-3,6])
grid()

axhline(lw=0.5)
axvline(lw=0.5)
legend()
plt.savefig('../plots/lgs_geom.pgf', bbox_inches='tight');
```
