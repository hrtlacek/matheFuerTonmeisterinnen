```python
import sympy as sp
from sympy.abc import x
```

```python
f = sp.Heaviside(x)*sp.exp(-x)
sp.plot(f)
```

```python
f.rewrite(sp.Piecewise)
```

```python
from sympy.abc import k
```

```python
F = sp.fourier_transform(f,x,k)
F
```

```python
sp.Abs(F)
```

```python
sp.arg(F)
```

```python
sp.plot(20*sp.log(sp.Abs(F)))
```
