```python
%pylab inline
import scipy.signal as sig
```

```python
sr = 44100
T = 3
N = int(sr*T)
t = arange(N)/sr
```

```python
f = 1000
x = sin(2*pi*f*t)
```

```python
faxis, mag = sig.welch(x, fs=sr, nperseg=2**12)
```

```python
figure(figsize=[6,6])
plot(faxis, mag, 'k-')
plt.savefig("../plots/specPlain.pgf", bbox_inches='tight') 
```

```python
figure(figsize=[6,6])
semilogx(faxis, 10*log10(mag), 'k-')
ylabel('Mag [dB]')
xlabel('Freq. [Hz]')
plt.savefig("../plots/specPlainLoglog.pgf", bbox_inches='tight') 
```
