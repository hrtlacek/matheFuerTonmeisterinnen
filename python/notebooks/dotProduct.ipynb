```python
sr = 1000
T = 3
 = pi
t = arange(int(T*sr))/sr
f = 50 #Hz
x1 = sin(2**f*t)
x2 = sin(2**f*t)

nCycles = 5

figure(figsize=[6.5,3])
ax = subplot(131)
plot(t,x1,'k', label=r'$a = sin(2\pi 50t)$')
plot(t,x2, '--r', label=r'$b = sin(2\pi 50t)$')
title(f'$dot(a,b)=${dot(x1,x2):.2f}')
xlim([0,nCycles/f])
axhline(lw=0.5)
legend(loc=3)
xlabel('$t$')

subplot(132,sharey=ax)
plot(t,x1, label=r'$a = sin(2\pi 50t)$')
plot(t,-x2, '--r', label=r'$b = -sin(2\pi 50t)$')
plt.gca().get_yaxis().set_visible(False)
title(f'$dot(a,b)=${dot(x1,-x2):.2f}')
xlim([0,nCycles/f])
axhline(lw=0.5)
legend(loc=3)
xlabel('$t$')

subplot(133,sharey=ax)
plot(t,x1, label=r'$a = sin(2\pi 50t)$')
x2 = 2*(np.random.random(int(T*sr))-0.5)
plot(t,x2, 'r--', label='$b =$ noise')

plt.gca().get_yaxis().set_visible(False)
title(f'$dot(a,b)=${dot(x1,x2):.2f}')
xlim([0,nCycles/f])
axhline(lw=0.5)
xlabel('$t$')
legend(loc=3)
plt.savefig('../plots/dotProd.pgf', bbox_inches='tight', transparent=True);
```
