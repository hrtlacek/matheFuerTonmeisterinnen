```python
sr = 44100
T = 2
first = 0.5
lag = 0.5
t = arange(T*sr)/sr
x1 = sinc((t-first)*2*2*pi)
x2 = sinc((t-first-lag)*2*2*pi)

plot(t, x1)
plot(t,x2)
```

```python
corr = correlate(x1,x2,mode='full')
lags = np.arange(-len(x1) + 1, len(x1)) / sr  # Convert sample lag to time lag
```

```python
plot(lags,corr)
```

```python
estimatedDelay = lags[argmax(corr)]
```

```python
print(f"Estimted Delay: {abs(estimatedDelay):.5f} seconds")
print(f"Actual Delay: {lag:.5f} seconds")
print(f"Error: {(lag+estimatedDelay)*1000:.4f} milliseconds")
```
