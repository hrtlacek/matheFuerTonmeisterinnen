```python
import sympy as sp
```

```python
from sympy.abc import x
```

```python
f = sp.Mod(x, 1)
f = x**2
```

```python
T = 2
f = sp.Piecewise((x,x<=T/4),(T/2-x,(x>T/4) & (x<3*T/4)),(x-T,x>3*T/4))*2
```

```python
f
```

```python
sp.plot(f,(x,0,T))
```

```python
fs = sp.fourier_series(f, limits=(0,2))
```

```python
s = fs.truncate(n=3)
```

```python
s
```

```python
sp.plot(f,s,(x,0,T))
```
