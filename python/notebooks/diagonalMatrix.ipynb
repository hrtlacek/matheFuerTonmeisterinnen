## Scaling Property

```python
nPoints = 5
X = random.random([2,nPoints])-0.5
scatter(*X)
xlim([-1,1])
ylim([-1,1])
axhline()
axvline()
```

```python

A = diag([2,1,2,1,1])
sp.Matrix(A)
```

```python
Y = X@A
```

```python
figure(figsize=[6,6])
scatter(*X)
scatter(*Y, c='r')

plot(*X)
plot(*Y, 'r--')

xlim([-1,1])
ylim([-1,1])
axhline(lw=0.5)
axvline(lw=0.5)

# plt.savefig("../plots/rotMat.pgf", bbox_inches='tight') 
```
